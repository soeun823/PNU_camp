```python
import copy

# ğŸ”µ ê³ ê° ê¸°ë³¸ í´ë˜ìŠ¤
class Customer:
    def __init__(self, name, birthdate, grade, address, purchase_history):
        self._info = {
            "name": name,
            "birthdate": birthdate,
            "grade": grade,
            "address": address,
            "purchase_history": purchase_history  # ë¦¬ìŠ¤íŠ¸ë¡œ ê´€ë¦¬
        }

    def get_info(self):
        # ê³ ê° ì •ë³´ ì¶œë ¥
        return self._info

    def get_snapshot(self):
        # ê³ ê° ì •ë³´ë¥¼ ì–•ì€ ë³µì‚¬ë¡œ ì €ì¥
        return copy.deepcopy(self._info)  # ê¹Šì€ ë³µì‚¬ë¡œ ë³€ê²½ (ì¤‘ì²© ë°ì´í„° ë³´í˜¸)

    def rollback(self, snapshot):
        # ìŠ¤ëƒ…ìƒ·ì„ ì´ìš©í•´ ì •ë³´ ë³µì›
        self._info = copy.deepcopy(snapshot)  # ê¹Šì€ ë³µì‚¬ë¡œ ë³€ê²½

    def __str__(self):
        # ê°ì²´ ì¶œë ¥ ì‹œ ì •ë³´ê°€ ë³´ì´ë„ë¡ ì¶”ê°€
        return str(self._info)


# ğŸŸ¢ ì¼ë°˜ ê³ ê° (ê°€ë³€ ê°ì²´)
class RegularCustomer(Customer):
    def update_info(self, key, new_value):
        # keyê°€ ìˆìœ¼ë©´ ê°’ ìˆ˜ì •
        if key in self._info:
            self._info[key] = new_value
        else:
            print(f"'{key}' í•­ëª©ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")


# ğŸŸ¡ VIP ê³ ê° (ë¶ˆë³€ ê°ì²´)
class VIPCustomer(Customer):
    def update_info(self, key, new_value):
        # VIP ê³ ê°ì€ ì •ë³´ë¥¼ ìˆ˜ì •í•  ìˆ˜ ì—†ìœ¼ë¯€ë¡œ, ìƒˆ ê°ì²´ë¥¼ ë°˜í™˜
        if key in self._info:
            # ê¸°ì¡´ ì •ë³´ë¥¼ ë³µì‚¬ í›„ ìƒˆ ê°ì²´ ë°˜í™˜
            new_info = copy.deepcopy(self._info)
            new_info[key] = new_value
            return VIPCustomer(**new_info)
        else:
            print(f"'{key}' í•­ëª©ì´ ì¡´ì¬í•˜ì§€ ì•ŠìŠµë‹ˆë‹¤.")
            return self  # ë³€ê²½ ë¶ˆê°€ ì‹œ ìê¸° ìì‹  ë°˜í™˜


if __name__ == "__main__":
    print("=== ê³ ê° ê´€ë¦¬ ì‹œìŠ¤í…œ ê¸°ëŠ¥ í…ŒìŠ¤íŠ¸ ===")

    # ì´ˆê¸° ì •ë³´ (ì¤‘ì²© ìë£Œ í¬í•¨ - purchase_history)
    vip_info = {
        "name": "Alice",
        "birthdate": "1990-01-01",
        "grade": "VIP",
        "address": "Seoul",
        "purchase_history": [{"item": "Laptop", "price": 1500}]
    }

    regular_info = {
        "name": "Bob",
        "birthdate": "1985-05-12",
        "grade": "Regular",
        "address": "Busan",
        "purchase_history": [{"item": "Smartphone", "price": 800}]
    }

    # VIP ê³ ê° (ë¶ˆë³€)
    vip = VIPCustomer(**vip_info)
    print("ì´ˆê¸° VIP ê³ ê°:", vip)

    # ìŠ¤ëƒ…ìƒ· ì €ì¥
    vip_snapshot = vip.get_snapshot()

    # Regular ê³ ê° (ê°€ë³€)
    reg = RegularCustomer(**regular_info)
    print("ì´ˆê¸° Regular ê³ ê°:", reg)

    reg_snapshot = reg.get_snapshot()

    # ì—…ë°ì´íŠ¸ ì‹œë„
    # VIP ê³ ê° ì—…ë°ì´íŠ¸: ìƒˆ ê°ì²´ ë°˜í™˜
    vip_updated = vip.update_info("address", "Incheon")

    # Regular ê³ ê° ì—…ë°ì´íŠ¸: in-place ë³€ê²½
    reg.update_info("address", "Daegu")

    # êµ¬ë§¤ ì´ë ¥ ìˆ˜ì •: ì¤‘ì²© ìë£Œ ìˆ˜ì • íš¨ê³¼ í™•ì¸
    vip_updated.get_info()["purchase_history"][0]["price"] = 1400
    reg.get_info()["purchase_history"][0]["price"] = 750

    print("\nì—…ë°ì´íŠ¸ í›„ VIP ê³ ê°:")
    print("ì›ë³¸:", vip)  # ì›ë³¸ì€ ë³€í™” ì—†ìŒ (ë¶ˆë³€)
    print("ì—…ë°ì´íŠ¸ëœ VIP:", vip_updated)
    print("VIP ìŠ¤ëƒ…ìƒ·:", vip_snapshot)

    print("\nì—…ë°ì´íŠ¸ í›„ Regular ê³ ê°:")
    print("í˜„ì¬ ìƒíƒœ:", reg)
    print("Regular ìŠ¤ëƒ…ìƒ·:", reg_snapshot)

    # ì¤‘ì²© ìë£Œê°€ mutableì´ë¯€ë¡œ, regular ìŠ¤ëƒ…ìƒ·(ì–•ì€ ë³µì‚¬)ëŠ” ê°™ì´ ë³€í–ˆì„ ìˆ˜ ìˆìŒ.
    # ë¡¤ë°± í…ŒìŠ¤íŠ¸
    reg.rollback(reg_snapshot)
    print("\nRegular ê³ ê° ë¡¤ë°± í›„:", reg)

    # VIP ê³ ê° ë¡¤ë°±: ìƒˆ ê°ì²´ ë°˜í™˜
    vip_rolled_back = vip_updated.rollback(vip_snapshot)
    print("\nVIP ê³ ê° ë¡¤ë°± í›„:", vip_rolled_back)
```
